# Secara Visual

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn')
```

```python
iris = pd.read_csv('FB_data.csv')
```

```python
iris.shape
```

```python
iris.head()
```

```python
iris.plot(x='High', y='Low')
```

```python
iris.plot(x='High', y='Low', kind='scatter')
plt.xlabel('High')
plt.ylabel('Low')
```

```python
iris.plot(y='High', kind='box')
plt.ylabel('Low ')
```

```python
iris.plot(y='High', kind='hist')
plt.xlabel('high')
```

```python
iris.plot(y='High', kind='hist', bins=30, range=(150,250))
plt.xlabel('high ')
plt.show()
```

```python
iris.plot(y='High', kind='hist', bins=30, range=(150,250), cumulative=True)
plt.xlabel('high ')
plt.title('Cumulative distribution function (CDF)')
```
